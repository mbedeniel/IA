L'objectif de ce BE est de réaliser une calculette commandable par la voie. Ainsi, la première étape est la construction des modèles acoustico-phonétiques nécessaires au décodage des mots à partir du signal acoustique. Nous utiliserons les modèles de Markov cachée (HMM). 

Premièrement, nous présenterons la topologie des modèles que nous proposons. Puis, l'évaluation de chacun de ses modèles. Une fois les modèles effectifs, il s'agira d'établir la grammaire nécessaire pour notre calculette.

## Topologie des modèles

Nous avons raisonné en termes de nombre phonèmes. ie, pour chaque mot, nous déterminons le nombre de phonèmes qui le constitue. Puis, nous déterminons les différentes prononciations possibles de ce mot à l'oral. Par exemple a l'oral certain prononcerons `voiture` en insistant sur le `e` à l'inverse, d'autres ne lirons pas le `e`. Une fois les différentes prononciations déterminées, nous constituons le modèle de Markov caché du mot correspondant. Un modèle de Markov caché est constitué d'un état de début(D), d'un état fin(F) et d'un ensemble d'état émetteur. Nous supposerons que chaque état émetteur aura pour charge de reconnaitre un phonème. Ainsi pour un mot de 3 phonèmes, nous aurons 5 états dont 3 états émetteurs. 
Dans la suite, nous donnerons pour chaque mot les différentes prononciations envisagé, les nombres de phonèmes associés et le modèle correspondant. Il faut noter que les poids sur les arcs ne sont que pour l'initialisation. Le modèle apprendra les poids corrects.

## Résultats de l'évaluation des modèles

On atteint une accuracy de **`97.50%`**. La matrice de confusion révèle qu'il y a une confusion entre les couples : **(un, cinq)** et **(un, quatre)**. En observant le **modèle de cinq** et celui de **un**, on note une certaine ressemblance des modèles (même nombre d'états émetteurs dans certain cas) et de la prononciation des 2 mots. En particulier si on compare le phonème **/ɛ̃/** de **cinq** à **un**. Le second modéle de **un** que nous proposons est beaucoup plus proche de la logique que nous utilisons (nombre d'états émetteur = nombre de phonèmes). De plus, avec ce deuxième modèle de **un**, on note moins de ressemblance avec le modèle de **cinq**.
